```python
import urllib.request

#사용할 pdf를 /pdfs 폴더 아래에 저장
#이 PDF는 삼성(Samsung) 스마트폰 모델 SM-S921W, SM-S926W, SM-S928W에 대한 사용자 가이드
urllib.request.urlretrieve(
    "https://downloadcenter.samsung.com/content/UM/202402/20240220134027613/S92X_UG_CA_UU_ENG_D9.pdf",
    filename="./pdfs/S92X_UG_CA_UU_ENG_D9.pdf"
)
```

```python
from langchain_community.embeddings import OpenAIEmbeddings
#openai 활용, 단어 임베딩 생성

from langchain.document_loaders import PyPDFLoader
#PDF파일 로드 하기 위함

from langchain.document_loaders import PyPDFDirectoryLoader
# 폴더(directory)내 여러 개 pdf 파일 로드

from langchain.text_splitter import CharacterTextSplitter
#텍스트를 문자 단위로 분할

from langchain_chroma import Chroma

folder_path = "./pdfs"
texts = []

#1000개의 문자를 담을 수 있는 박스(chunk), 마지막 50자는 겹치게
text_splitter = CharacterTextSplitter(separator="\n",
                    chunk_size = 1000,
                    chunk_overlap=50)

for filename in os.listdir(folder_path):
    if filename.endswith(".pdf"):
        raw_documents = PyPDFLoader(folder_path + "/" + filename).load()
        documents = text_splitter.split_documents(raw_documents)
        texts.extend(documents)

print("*****")
print(len(texts))
print(texts)
print("*****")
```

```python
texts[100].page_content
```

```python
db = Chroma.from_documents(texts, OpenAIEmbeddings())
```

```python
retriever = db.as_retriever(search_kwargs = {'k' : 10})
```

```python
import os
import re

from operator import itemgetter

from langchain_core.prompts import ChatPromptTemplate

from langchain.schema.output_parser import StrOutputParser

from langchain.schema.runnable import RunnablePassthrough
from langchain_openai import ChatOpenAI
```

```python
template = \
"""너는 훌륭한 비서이다. 아래 [질문]에 대해서 [내용]을 참고하여 답변을 한국어로 작성한다. meta 데이터도 표시한다. \n\n
[내용] -- 시작 --
{context}
[내용] -- 끝 --
[질문]
{question}
"""


```

```python
chat_prompt_template = ChatPromptTemplate.from_template(template)
chat_model = ChatOpenAI(model_name = 'gpt-4o')

def format_docs(docs):
    return "\n\n".join(doc.page_content for doc in docs)

chain = (
    {'context' : retriever, 'question' : RunnablePassthrough()}
    | chat_prompt_template
    | chat_model
    | StrOutputParser()
)

#메타데이터 없이 답변만 만들려면 retriever | format_docs
```

```python
def chat_with_user(user_message):
    ai_message = chain.invoke(user_message)
    return ai_message
#계속 답변을 받아온다.
while True:
    user_message = input("USER > ")
    #quit를 받을 때 까지 반복해서 유저의 입력을 받는다.
    if user_message.lower() == "quit":
        break
    #quit가 들어오지 않는 한 채팅을 통해 답변을 받는다.
    ai_message = chat_with_user(user_message)
    print(f" A I > {ai_message}")
```
